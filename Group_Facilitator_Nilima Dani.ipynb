```python
#Convert term column from string to int
def clean_term(t):
    return int(t.strip().split()[0])
    
data['term'] = data['term'].apply(clean_term)
data['term'].dtype
```

```python
#when column value is 1. deletting them as well.
for col in data.columns:
    if data[col].nunique(dropna=True) == 1:
        print(col)
        data = data.drop(col, axis=1)

print("--------------")
data
```

```python
#Drop the column which is having more than 30% of value is null.
tot_row = len(data)
for col in data.columns:
    if data[col].isna().mean().round(2) >= 0.30:
        print(data)
        data = data.drop(col, axis=1)
        
data.shape
```

```python
# Assuming you have already computed stat_vs_loan DataFrame

# Grouping and calculating proportions
stat_vs_loan = data.groupby(['addr_state', 'loan_status']).size().unstack(fill_value=0).reset_index()
stat_vs_loan['Total'] = stat_vs_loan['Charged Off'] + stat_vs_loan['Fully Paid']
stat_vs_loan['Chargedoff_Proportion'] = stat_vs_loan['Charged Off'] / stat_vs_loan['Total']
stat_vs_loan = stat_vs_loan.sort_values('Chargedoff_Proportion', ascending=False)

# Plotting
fig, ax1 = plt.subplots(figsize=(12, 8))  # Create a figure and axis
fig.patch.set_facecolor('blue')  # Set figure facecolor directly

ax1.set_title('Charged-off Proportion by State', fontsize=14, color='white')
ax1 = sns.barplot(x='addr_state', y='Chargedoff_Proportion', data=stat_vs_loan, ax=ax1, palette='viridis')
ax1.set_xlabel('State', fontsize=14, color='white')
ax1.set_ylabel('Charged-off Proportion', fontsize=14, color='white')
ax1.tick_params(axis='x', rotation=45, labelcolor='white')  # Rotate x-axis labels and set color

plt.tight_layout()  # Ensure tight layout to prevent clipping of labels
plt.show()
```

```python
#Finding correlation matric
data = data.select_dtypes(include=[float, int])
corr_matrix = data.corr()
plt.figure(figsize=(20,20))
#plotting correlatioon matric on a heat map
ax = sns.heatmap(corr_matrix, annot = True, cmap='Blues')
top, bottom = ax.get_ylim()
ax.set_ylim(top+0.5, bottom-0.5)
plt.title("Correlation between Columns")
plt.show()
```
